# LSG50 Index — Project Architecture

Este notebook describe la **arquitectura completa del proyecto LSG50**, incluyendo el flujo de datos, la organización de notebooks y la estructura del pipeline.

El objetivo es documentar de forma clara cómo se construye el índice desde la extracción de datos hasta el análisis final.

## Project Overview

El proyecto tiene como objetivo diseñar y evaluar un **índice de renta variable basado en fundamentales financieros**, construido a partir del universo de empresas del S&P 500.

El pipeline completo incluye:

1. Extracción de datos financieros desde una API
2. Limpieza y transformación de datos
3. Construcción del índice LSG50
4. Backtest histórico del portafolio
5. Análisis y documentación del índice

## Notebook Structure

El proyecto está organizado en varios notebooks que representan distintas etapas del pipeline.

| Notebook | Función |

## LSG50 Index Methodology

El índice LSG50 se construye utilizando un ranking propietario denominado **CFO Index**.

Este indicador combina distintas métricas fundamentales para evaluar la fortaleza financiera de cada empresa.

## Backtest Methodology

Para evaluar el comportamiento del índice se realiza un backtest utilizando precios históricos descargados desde Yahoo Finance.

El análisis incluye:

- cálculo de retornos diarios
- construcción del retorno del portafolio
- comparación con el benchmark S&P 500
- cálculo de métricas de desempeño

## Project Goals

Este proyecto tiene varios objetivos:

• Aplicar técnicas de análisis de datos en un contexto financiero  
• Diseñar un índice cuantitativo basado en fundamentales  
• Evaluar el desempeño histórico del índice  
• Construir un pipeline reproducible de análisis financiero  

El proyecto combina herramientas como **Python, SQL, APIs financieras y análisis cuantitativo**.
