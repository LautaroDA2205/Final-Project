```python
import os
import requests
import json
from dotenv import load_dotenv

# Cargar variables de entorno desde .env
import os
print("Working directory:", os.getcwd())

print("Env file exists:", os.path.exists("../../.env"))
load_dotenv("../../.env", override=True)
print("Valor real:", os.environ.get("FMP_API_KEY"))

API_KEY = os.getenv("FMP_API_KEY")

print("API Key cargada:", API_KEY is not None)

ticker = "AAPL"

url = f"https://financialmodelingprep.com/api/v3/income-statement/{ticker}?limit=10&apikey={API_KEY}"

response = requests.get(url)

print("Status Code:", response.status_code)

if response.status_code == 200:
    data = response.json()
    os.makedirs("../data_raw", exist_ok=True)
    
    with open(f"../data_raw/{ticker}_income.json", "w") as f:
        json.dump(data, f, indent=4)
    
    print("Archivo guardado correctamente.")
else:
    print("Error en la solicitud.")
```
